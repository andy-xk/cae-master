```python
from collections import OrderedDict
from fwf import BaseLineParser, BaseFileParser


class bdfLineParser(BaseLineParser):
    """Complete line parser for humans.txt example file."""

    _map = OrderedDict(
        [
            ("keyword", slice(0, 8)),
            ("idx", slice(9, 16)),
            ("field2", slice(17, 24)),
            ("field3", slice(25, 32)),
            ("field4", slice(33, 40)),
            ("field5", slice(41, 48)),
            ("field6", slice(49, 56)),
            ("field7", slice(57, 64)),
            ("field8", slice(65, 72)),
        ]
    )


class BdfFileParser(BaseFileParser):
    """Complete file parser for humans.txt example file."""

    _line_parser = bdfLineParser
```

```python
bdf = BdfFileParser.open(path)
```
